```python
data= pd.read_csv("https://media.geeksforgeeks.org/wp-content/uploads/nba.csv")
```

Fill na  values

```python
data1 = data.fillna(value=0)
data1.head()
```

```python
# fill null value with the mean of a column
df= data.fillna(value=data['College'].mean) 
```

drop nan row

```python
data4 = data.dropna()
data4.head()
```

# replace

```python
import numpy as np
data5 = data.replace(to_replace=np.nan, value=8988)
```

# interpolate

```python
data6['College']= data['College'].interpolate(method='linear' )
```

 concatenating team with name column  
 overwriting name column

```python
new_data = data["Name"].copy()
data['New_name'] = data['Name'].str.cat(new_data, sep=",")
```

```python
# making copy of team column
new = data["Team"].copy()
 
# string to replace null values with
na_string ="No College"
 
# concatenating team with name column
# overwriting name column
data["College"]= data["College"].str.cat(new, sep =", ", na_rep = na_string)
```
