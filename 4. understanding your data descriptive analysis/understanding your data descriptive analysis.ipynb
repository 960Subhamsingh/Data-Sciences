count how many people are  travel in train with male and female

```python
df['Sex'].value_counts().plot(kind='pie', autopct='%.2f'  , legend=['Male' ,'Female'] , title='count of sex'  )
```

```python
male=df[df['Sex']=='male']
male.Sex.value_counts()
```

```python
female= df[df['Sex']=='female']
female.Sex.value_counts()
```

```python
plt.scatter( female.Fare, female.Age,
label='Female',color='orange',marker='*')

plt.scatter(male.Fare, male.Age , label='male', color='purple' , marker='+')

plt.xlabel('Fare')
plt.ylabel('Age')
plt.title('Male V/s female')

plt.show()
```
