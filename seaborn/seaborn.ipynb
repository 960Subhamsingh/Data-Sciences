```python
import seaborn as sns
df = sns.load_dataset('tips')
print (df.head())
```

```python
d =df['smoker'].value_counts()
g = df['sex'].value_counts()
```

```python
sns.countplot(x='sex',data=df , hue='sex')
```

```python
sns.boxplot(x='day', y='total_bill' ,data=df,hue='smoker') #categorical  data
```

```python
sns.violinplot(x='day',y='total_bill',data=df,hue = 'sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=df , hue='day')
```

```python
sns.barplot(x='sex', y='total_bill', data=df, hue='sex')
```

```python
sns.distplot(df['total_bill'])
```

```python
sns.distplot(df['total_bill'],kde = False,bins=30,color='r')
```

```python
sns.kdeplot(df['total_bill'])
```

```python
sns.jointplot(x='day',y='total_bill',data=df) #comparision
```

```python
sns.pairplot(df) #relationship
```

```python
sns.relplot(data = df , x="total_bill" , y="tip"  , col="time" , hue= "smoker"
            ,style="smoker" , size = "size"  ) 
```

```python
sns.pairplot(df,hue='day')
```

```python
df.corr() # correlation
```

```python
sns.heatmap(df.corr(),annot=True)
```

Distribution Plots
These plots help us to visualize the distribution of data. We can use these plots to understand the mean, median, range, variance, deviation, etc of the data.

a. Distplot
Dist plot gives us the histogram of the selected continuous variable.
It is an example of a univariate analysis.
We can change the number of bins i.e. number of vertical bars in a histogram

```python
sns.displot(x=data['Age'], bins=42)
```
