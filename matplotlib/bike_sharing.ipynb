Get a summary of the initial version of a dataframe

Convert to datetime

```python
df['datetime'] = pd.to_datetime(df['datetime'])
```

Get the weekday, month, and hour

```python
df['weekday'] = df['datetime'].dt.dayofweek
df['hour'] = df['datetime'].dt.hour
df['month'] = df['datetime'].dt.month
```

Convert certain columns to the 'category' data type
 

```python
categories = ['season', 'workingday', 'weekday', 'hour', 'month', 'weather', 'holiday']
for category in categories:
    df[category] = df[category].astype('category')
```

Substitute values of categorical variables

```python
 
df['season'] = df['season'].replace([1, 2, 3, 4], ['spring', 'summer', 'fall', 'winter'])
df['season'].unique() # check results of values substitution
 
```
