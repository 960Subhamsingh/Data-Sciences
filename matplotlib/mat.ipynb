```python
df = pd.DataFrame({"A": ["burger", "burger", "burger", "burger", "burger",
                         "fries", "fries", "fries", "fries"],
                   "B": ["one", "one", "one", "two", "two",
                         "one", "one", "two", "two"],
                   "C": ["small", "large", "large", "small",
                         "small", "large", "small", "small",
                         "large"],
                   "D": [1, 2, 2, 3, 3, 4, 5, 6, 7],
                   "E": [2, 4, 5, 5, 6, 6, 8, 9, 9]})
```

```python
pd.pivot_table(df, index=['A', 'B'], aggfunc=np.sum)
```

```python
pd.pivot_table(df, index=['A', 'B'], columns=['C'] ,aggfunc={'D':np.sum, 'E': np.mean})
```
